## Step 1: Storing Customer Orders

We store customer names and order details using lists and tuples.

```python
# List of customers
customers = ["Alice", "Bob", "Charlie", "David", "Eva"]
```

```python
# Orders → (customer, product, price, category)
orders = [
    ("Alice", "Laptop", 800, "Electronics"),
    ("Alice", "Shoes", 60, "Clothing"),
    ("Bob", "Phone", 500, "Electronics"),
    ("Bob", "T-shirt", 30, "Clothing"),
    ("Charlie", "Blender", 120, "Home"),
    ("Charlie", "Laptop", 900, "Electronics"),
    ("David", "Shirt", 40, "Clothing"),
    ("Eva", "Microwave", 150, "Home"),
    ("Eva", "Headphones", 80, "Electronics")
]
print("Orders Loaded Successfully")
```

##  Step 2: Organizing Data using Dictionary

```python
customer_orders = {}

for name, product, price, category in orders:
    if name not in customer_orders:
        customer_orders[name] = []
    customer_orders[name].append((product, price, category))

print(customer_orders)
```

##  Step 3: Product Category Classification

```python
product_category = {}

for name, product, price, category in orders:
    product_category[product] = category

# Unique categories
categories = set(product_category.values())

print("Categories:", categories)
```

##  Step 4: Customer Spending and Classification

```python
customer_spending = {}
customer_type = {}

for name, product, price, category in orders:
    customer_spending[name] = customer_spending.get(name, 0) + price

# Classification
for name, total in customer_spending.items():
    if total > 100:
        customer_type[name] = "High-value"
    elif 50 <= total <= 100:
        customer_type[name] = "Moderate"
    else:
        customer_type[name] = "Low-value"
```

##  Step 5: Business Insights Generation

```python
category_revenue = {}

for name, product, price, category in orders:
    category_revenue[category] = category_revenue.get(category, 0) + price

print("Revenue per category:", category_revenue)
```

```python
unique_products = set()

for _, product, _, _ in orders:
    unique_products.add(product)

print("Unique Products:", unique_products)
```

```python
electronics_customers = [
    name for name, product, price, category in orders
    if category == "Electronics"
]

print("Electronics buyers:", set(electronics_customers))
```

```python
top_customers = sorted(customer_spending.items(), key=lambda x: x[1], reverse=True)

print("Top 3 Customers:", top_customers[:3])
```

```python
customer_categories = {}

for name, product, price, category in orders:
    if name not in customer_categories:
        customer_categories[name] = set()
    customer_categories[name].add(category)

multi_category_customers = [name for name, cats in customer_categories.items() if len(cats) > 1]

print("Multi-category customers:", multi_category_customers)
```

##  Step 6: Set Operations on Customers

```python
electronics = set()
clothing = set()

for name, product, price, category in orders:
    if category == "Electronics":
        electronics.add(name)
    if category == "Clothing":
        clothing.add(name)

common_customers = electronics & clothing

print("Bought both Electronics & Clothing:", common_customers)
```

## Final Summary

```python
print("\n--- Customer Summary ---")
for name in customers:
    print(f"{name}: Spent = {customer_spending.get(name,0)}, Type = {customer_type.get(name)}")
```

##  Conclusion

- Electronics category generated the highest revenue.
- High-value customers contribute most to sales.
- Customers purchasing from multiple categories are more valuable.
- The analysis helps businesses improve marketing and inventory strategies.
